### If Statements

Python uses boolean variables to evaluate conditions. The boolean values True and False are returned when an expression is compared or evaluated.

The basic structure is:

```python
if condition:
    code_executed_if_true
```

Not that the next line (and all code in the code block that is to be executed if the condition is true) must be indented. This is how python tells what code is part of the if block (there are no curly braces or do-end type syntax).

Notice that variable assignment is done using a single equals operator "=", whereas comparison between two variables is done using the double equals operator "==". The "not equals" operator is marked as "!=". ">", "<", ">=", "<=" are used for greater, less, greater-than-equal, and less-than-equal for numeric data.

````python
my_int_a = 4
my_int_b = 5

if True:
    print "This is true"
    
if my_int_a % 2 == 0:
    print "a is even"
    
if my_int_a % 2 == 0 and my_int_b % 2 == 0:
    print "a and b is even"
    
if my_int_b % 2 == 0:
    print "b is even"
    if my_int_b < 3:
        print "b is less than 3"
elif my_int_b % 5 == 0:
    print "b is divisible by 5"
else:
    print "b isn't divisible by 2, 3, or 5"
````

### For/while loops

For loops iterate over a given sequence.

While loops repeat as long as a certain boolean condition is met. Break is used to exit a for loop or a while loop.

The basic structure is:

```python
for var in list:
    code_thats_executed_for_each_var
    
while condition:
    code_executed_while_true
```

As with if statements, the code in the for or while block must be indented consistently.

````python
i = 0
print "Loop 1:"
while i < 5:
    print i
    i = i + 1

print "\nLoop 2:"
i = 0
while True:
    if i >= 5:
        break
    print i
    i = i + 1
````

#### Group-by and statistics

The basic strategy for aggregating and getting group statistics is to use the groupby function by passing it a list of the variables that will make up the group.

You can then call various functions such as sum(), mean(), etc. that will will calculate the aggregate for the group. Alternatively you can use the agg() function by passing it a dictionary where the keys are the variables you want aggregate statistics of, and the values are a string or list of strings of the different functions you want to aggregate with.

The groupby() function's level= parameter is useful for merging a more detailed value to a more aggregated value and calculating things based on that.

````python
group_df = join_df.reset_index()
group_df = group_df[['DISTRICT_NAME', 'ANTICIPATED_DELV_DATE', 'ONTIME', 'WTEDPCS']]
group_df = group_df.sort_values(['DISTRICT_NAME', 'ANTICIPATED_DELV_DATE', 'ONTIME'])
print group_df.head(10)

group_df1 = group_df.groupby(['DISTRICT_NAME', 'ANTICIPATED_DELV_DATE', 'ONTIME']).sum()
group_df1.head(10)
````

The above code block is equivalent to this sas code:

```sas
proc summary data = group_df;
    class district_name anticipated_delv_date ontime;
    var wtedpcs;
    output out=group_df (drop = _type_ _freq_) sum=;
run;
```

or the following sql:
    
```sql
CREATE TABLE group_df AS
SELECT district_name, anticipated_delv_date, ontime, sum(wtedpcs) as wtedpcs
FROM group_df
GROUP BY district_name, anticipated_delv_date, ontime;
```

#### Supervised learning - classification with Decision Trees

Decision trees are a simple machine learning model to start off with. Scikit-learn has a [tree](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.tree) library for building and working with decision trees. This shows the basic recipe for defining a model, fitting a model based on training data. Calculating the training and test accuracy.

Other modules can be used to visualize the decision tree that was geneated.

When creating a model you can make your own choice for different "hyper-parameters" about the model. Using these hyper-parameters you can change to what extent the model "learns" from the training data to avoid overfitting.

````python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_train_pred = clf.predict(X_train)
print y_train_pred
train_accuracy = 1.0 * sum(y_train_pred == y_train.reshape(-1)) / y_train.size
print train_accuracy

y_pred = clf.predict(X_test)
test_accuracy = 1.0 * sum(y_pred == y_test.reshape(-1)) / y_test.size
print test_accuracy
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

knn = KNeighborsClassifier()
knn = knn.fit(X_train, y_train.reshape(-1))
y_train_pred = knn.predict(X_train)
train_accuracy = 1.0 * sum(y_train_pred == y_train.reshape(-1)) / y_train.size
print "knn train: %f" % train_accuracy

y_pred = knn.predict(X_test)
test_accuracy = 1.0 * sum(y_pred == y_test.reshape(-1)) / y_test.size
print "knn test: %f" % test_accuracy

##############################

rfc = RandomForestClassifier(n_estimators = 50)
rfc = rfc.fit(X_train, y_train.reshape(-1))
y_train_pred = rfc.predict(X_train)
train_accuracy = 1.0 * sum(y_train_pred == y_train.reshape(-1)) / y_train.size
print "random forest train: %f" % train_accuracy

y_pred = rfc.predict(X_test)
test_accuracy = 1.0 * sum(y_pred == y_test.reshape(-1)) / y_test.size
print "random forest test: %f" % test_accuracy

##############################

abc = AdaBoostClassifier(base_estimator = tree.DecisionTreeClassifier(max_depth = 5), n_estimators=50)
abc = abc.fit(X_train, y_train.reshape(-1))
y_train_pred = abc.predict(X_train)
train_accuracy = 1.0 * sum(y_train_pred == y_train.reshape(-1)) / y_train.size
print "Boosting train: %f" % train_accuracy

y_pred = abc.predict(X_test)
test_accuracy = 1.0 * sum(y_pred == y_test.reshape(-1)) / y_test.size
print "Boosting test: %f" % test_accuracy
````
